```python
""" Creamos una matriz de correlación entre las columnas del conjunto de datos y luego
visualiza esta matriz de correlación como un mapa de calor utilizando la biblioteca Seaborn y Matplotlib.
Aquí tienes una explicación de lo que hace cada parte del código:

1) corr = dataset.corr(): Esta línea calcula la matriz de correlación entre todas las columnas numéricas
del DataFrame dataset. La matriz de correlación muestra cómo las variables numéricas se relacionan entre sí.
Cada celda en la matriz contiene el coeficiente de correlación entre dos variables.

2) fig, ax = plt.subplots(figsize=(8, 6)): Aquí se crea una figura y un eje para la visualización.
Se especifica un tamaño de figura de 8x6 pulgadas.

3) sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f"): Esto crea un mapa de calor utilizando Seaborn.
El mapa de calor muestra la matriz de correlación con colores, donde los colores más cálidos indican una
correlación positiva y los colores más fríos indican una correlación negativa. Los argumentos
cmap='coolwarm' establecen la paleta de colores, annot=True agrega valores numéricos en las celdas y fmt=".2f"
formatea los valores con dos decimales.

4) plt.xticks(range(len(corr.columns)), corr.columns);: Esto establece las etiquetas en el eje x
para que coincidan con las columnas del DataFrame. Las etiquetas son los nombres de las columnas.

5) plt.yticks(range(len(corr.columns)), corr.columns): Similar al paso anterior,
esto establece las etiquetas en el eje y.

6) plt.show(): Finalmente, esto muestra el mapa de calor en la figura creada anteriormente."""

corr = df.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(8, 6))
#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```

```python
""" El código utiliza pd.get_dummies() de Pandas para crear variables dummy (variables binarias)
a partir de las columnas categóricas en el DataFrame dataset. Esto es útil cuando queremos convertir
variables categóricas en una representación numérica que sea adecuada para su uso en algoritmos
de machine learning que requieren entradas numéricas en lugar de categóricas.

1) df = pd.get_dummies(dataset): Esta línea aplica la función pd.get_dummies() al DataFrame dataset.
Esta función identifica automáticamente las columnas categóricas en el DataFrame y crea variables
dummy para cada categoría única en esas columnas. Básicamente, convierte las columnas categóricas
en múltiples columnas binarias, una para cada categoría única. El resultado se almacena en un nuevo DataFrame llamado df.

2) df.head(): Luego, df.head() se utiliza para mostrar las primeras filas del nuevo DataFrame df que contiene las variables dummy.

3) Después de ejecutar este código, df contendrá todas las columnas originales del DataFrame dataset,
junto con las nuevas columnas dummy generadas para las columnas categóricas. Esto permite que los modelos
de machine learning utilicen las variables categóricas como características en lugar de tener que lidiar con ellas de manera categórica. """

# Varibles dummies:
df = pd.get_dummies(df)
df.head()
```

```python
""" El código está dividiendo un DataFrame en dos conjuntos de datos, X y y,
que son comunes en el contexto de problemas de machine learning y análisis de datos:

X = df.drop('cantidad_casos', axis=1): En esta línea, se crea un nuevo DataFrame X al eliminar la
columna 'cantidad_casos' del DataFrame original df utilizando el método drop(). El argumento axis=1 se
utiliza para indicar que se debe eliminar una columna (en lugar de una fila). Como resultado, X
contendrá todas las columnas de df excepto 'cantidad_casos'. En otras palabras, X contendrá las
características o variables independientes que se utilizarán para predecir 'cantidad_casos'.

y = df['cantidad_casos']: En esta línea, se crea una Serie y que contiene la columna 'cantidad_casos'
del DataFrame original df. Esto significa que y contendrá los valores que se intentarán predecir o modelar,
que generalmente se conoce como la variable objetivo o dependiente. En este caso, parece que estás tratando
de predecir la cantidad de casos ('cantidad_casos') utilizando las otras características en X.

Una vez que hayas ejecutado estas líneas de código, tendrás dos conjuntos de datos separados:
X, que contiene las características, e y, que contiene la variable objetivo.
Estos conjuntos de datos se pueden utilizar para entrenar modelos de machine learning y realizar análisis estadísticos. """

# Separamos el dataset:
X = df.drop('cantidad_casos', axis = 1)
y = df['cantidad_casos']
```

```python
""" El código divide el conjunto de datos en conjuntos de entrenamiento y
prueba utilizando la función train_test_split de Scikit-Learn.

from sklearn.model_selection import train_test_split: Esto importa la función train_test_split
del módulo sklearn.model_selection, que se utiliza para dividir el conjunto de datos en conjuntos de entrenamiento y prueba.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42): Esta línea divide los datos en cuatro conjuntos:

X_train: Contiene las características de entrenamiento.
X_test: Contiene las características de prueba.
y_train: Contiene los valores objetivo correspondientes a los datos de entrenamiento.
y_test: Contiene los valores objetivo correspondientes a los datos de prueba.
Los argumentos utilizados en train_test_split son los siguientes:

X: El conjunto de características (variables independientes) que deseas dividir.

y: El conjunto de valores objetivo (variable dependiente) que deseas dividir.

test_size: Indica el tamaño del conjunto de prueba. En este caso, se utiliza un valor del 20%,
lo que significa que el 20% de los datos se utilizarán como conjunto de prueba y el 80%
restante se utilizará como conjunto de entrenamiento.

random_state: Se utiliza para establecer una semilla aleatoria. Esto asegura que la división se realice
de la misma manera cada vez que se ejecute el código, lo que facilita la reproducibilidad de los resultados.
En resumen, después de ejecutar este código, tendrás los datos divididos en conjuntos de
entrenamiento (X_train y y_train) y conjuntos de prueba (X_test y y_test). Estos conjuntos se utilizan
para entrenar y evaluar modelos de machine learning respectivamente. """

# Splitting the dataset into
# the Training set and Test set:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state= 42) # 80% training
```

```python
print("Shape X_train: ",X_train.shape)
print("Shape X_test: ", X_test.shape)
print("Shape y_train: ",y_train.shape)
print("Shape y_test: ",y_test.shape)
```

```python
""" El código crea un modelo de regresión lineal utilizando la clase LinearRegression de Scikit-Learn.

regressor_linear = LinearRegression(): En esta línea, se crea una instancia de la clase LinearRegression y
se asigna a la variable regressor_linear. Esta instancia representa el modelo de regresión lineal que
utilizarás para hacer predicciones.

La regresión lineal es un algoritmo de machine learning utilizado para modelar relaciones lineales
entre una variable dependiente (objetivo) y una o más variables independientes (características).
Una vez que hayas creado el modelo de regresión lineal, podemos entrenarlo con los datos de entrenamiento y
luego utilizarlo para hacer predicciones sobre nuevos datos. """

# Creating the linear regressor
regressor_linear = LinearRegression()
```

```python

""" El código entrena el modelo de regresión lineal (regressor_linear) utilizando los
conjuntos de entrenamiento X_train (características) e y_train (valores objetivo).

regressor_linear.fit(X_train, y_train): Esta línea utiliza el método fit del modelo de regresión
lineal (regressor_linear) para entrenar el modelo utilizando los datos de entrenamiento.
El argumento X_train representa las características de entrenamiento y el argumento y_train
representa los valores objetivo correspondientes.Después de ejecutar esta línea de código,
el modelo de regresión lineal estará ajustado a los datos de entrenamiento y habrá aprendido una relación
lineal entre las características y los valores objetivo.
Esto significa que el modelo podrá hacer predicciones sobre nuevos datos basados en esta relación aprendida. """

# Entrenamos el modelo:
regressor_linear.fit(X_train, y_train)
```

```python
""" El código imprime los coeficientes del modelo de regresión lineal (regressor_linear).

regressor_linear.coef_: Esta línea accede al atributo coef_ del modelo de regresión lineal regressor_linear.
Los coeficientes representan las ponderaciones o contribuciones de cada característica (variable independiente)
en la predicción de los valores objetivo.

Al imprimir regressor_linear.coef_, obtenemos una lista de los coeficientes correspondientes a cada característica
en el modelo. Cada coeficiente indica cuánto afecta esa característica a la predicción del valor objetivo.
Por ejemplo, si tenemos una característica llamada "X1" y el coeficiente correspondiente es 2.5,
significa que un aumento de 1 unidad en "X1" se asocia con un aumento de 2.5 unidades en la variable objetivo.

Estos coeficientes son importantes para comprender la relación entre las características y los valores objetivo
en el modelo de regresión lineal y pueden ayudarnos a interpretar cómo cada característica influye en las predicciones. """

# Ahora podemos obtener los coeficientes:
print('Coeficientes:', regressor_linear.coef_)
```

```python
"""
El código imprime el coeficiente de intercepción (a menudo denotado como "b0" o "intercepto")
del modelo de regresión lineal (regressor_linear).

regressor_linear.intercept_: Esta línea accede al atributo intercept_ del modelo de regresión lineal regressor_linear.
El coeficiente de intercepción representa el valor estimado de la variable objetivo cuando todas las características
son iguales a cero. En otras palabras, es el valor "base" que se agrega a las predicciones en ausencia de influencia
de las características.

Al imprimir regressor_linear.intercept_, obtenemos el valor numérico del coeficiente de intercepción del modelo.
Este valor es importante para la ecuación de regresión lineal, que se puede expresar como:

y = b0 + b1*x1 + b2*x2 + ... + bn*xn

Donde:

y es la variable objetivo.
b0 es el coeficiente de intercepción.
b1, b2, ..., bn son los coeficientes de las características (x1, x2, ..., xn).

El coeficiente de intercepción determina dónde la línea de regresión corta el eje vertical (eje y)
cuando todas las características son iguales a cero. Esto es fundamental para la interpretación y
predicción de modelos de regresión lineal."""

# Ahora podemos obtener el coeficiente intercepto:
print('Coeficiente b0:', regressor_linear.intercept_)
```

```python
""" El código crea un DataFrame que contiene los coeficientes del modelo de
regresión lineal (regressor_linear) junto con el nombre de las características correspondientes.
pd.DataFrame(...): Se utiliza para crear un nuevo DataFrame de Pandas.

regressor_linear.coef_: Accede a los coeficientes del modelo de regresión lineal entrenado.

X.columns: Accede a los nombres de las características en el DataFrame X, que se utilizan como índice en el nuevo DataFrame.

columns=['coefficient']: Asigna el nombre de columna "coefficient" al nuevo DataFrame para identificar los coeficientes.

El resultado es un DataFrame que tiene dos columnas: una columna con los nombres de las características (X.columns)
y otra columna con los coeficientes correspondientes. Esto proporcionará una representación tabular de los
coeficientes y sus respectivas características, lo que puede ser útil para el análisis y la interpretación del modelo de regresión lineal."""

# Generamos un df con los coeficientes obtenidos:
pd.DataFrame(regressor_linear.coef_, X.columns, columns=['coefficient'])
```

```python
""" El código importa tres métricas comunes utilizadas para evaluar el rendimiento de modelos de regresión en Scikit-Learn:

mean_squared_error: Esta métrica calcula el error cuadrático medio (MSE) entre las predicciones del modelo y los valores reales.
El MSE mide la magnitud promedio de los errores al cuadrado, lo que significa que penaliza los errores más grandes de manera
más significativa. Un valor bajo de MSE indica un mejor ajuste del modelo a los datos.

r2_score: Esta métrica calcula el coeficiente de determinación (R²), que es una medida de cuánta variabilidad en los valores
objetivo es explicada por el modelo. R² varía entre 0 y 1, donde 1 indica una explicación perfecta de la variabilidad y
0 indica que el modelo no explica nada de la variabilidad.

mean_absolute_error: Esta métrica calcula el error absoluto medio (MAE) entre las predicciones del modelo y los valores reales.
El MAE mide la magnitud promedio de los errores sin elevarlos al cuadrado. Es una medida más robusta que el MSE frente a
valores atípicos.

Estas métricas son útiles para evaluar la precisión y el rendimiento general de modelos de regresión. Puedes usarlas después
de hacer predicciones con el modelo para determinar cuán bien se ajusta a los datos de prueba y si es necesario realizar
ajustes o mejoras en el modelo. """

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
""" El código calcula varias métricas para evaluar el rendimiento del modelo de regresión lineal
en función de las predicciones (y_pred) y los valores reales (y_test).
mean_squared_error(y_test, y_pred): Esta línea calcula el Error Cuadrático Medio (MSE)
entre los valores reales (y_test) y las predicciones del modelo (y_pred). El MSE mide la magnitud
promedio de los errores al cuadrado. Cuanto menor sea el valor del MSE, mejor será el ajuste del modelo a los datos.

mean_squared_error(y_test, y_pred, squared=False): Esta línea calcula la Raíz del Error Cuadrático Medio (RMSE),
que es la raíz cuadrada del MSE. El RMSE proporciona una medida de la magnitud promedio de los errores en la
misma unidad que la variable objetivo. Es una medida de dispersión que se interpreta de manera similar
al desviación estándar.

mean_absolute_error(y_test, y_pred): Esta línea calcula el Error Absoluto Medio (MAE) entre los valores
reales y las predicciones. El MAE mide la magnitud promedio de los errores sin elevarlos al cuadrado.
Es una métrica más robusta frente a valores atípicos que el MSE.

r2_score(y_test, y_pred): Esta línea calcula el coeficiente de determinación (R²) entre los valores reales
y las predicciones. El R² es una medida de cuánta variabilidad en los valores objetivo es explicada por el
modelo. Un valor de R² cercano a 1 indica que el modelo explica bien la variabilidad, mientras que un valor
cercano a 0 indica que el modelo no explica la variabilidad.

Al imprimir estas métricas, puedes evaluar el rendimiento del modelo de regresión lineal y determinar
cuán bien se ajusta a los datos de prueba. Generalmente, un MSE y un RMSE más bajos, un MAE más bajo y
un R² más cercano a 1 indican un mejor rendimiento del modelo. """

# Calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y_test, y_pred))
print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred,squared=False)))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```

```python
""" El código ajusta un modelo de regresión lineal utilizando el módulo ols de statsmodels en Python.

model = ols(formula="cantidad_casos ~ semanas_epidemiologicas + provincia_id + departamento_id + semanas_epidemiologicas + grupo_edad_id", data=df):
En esta línea, defines un modelo de regresión lineal utilizando la función ols de statsmodels.
La fórmula "cantidad_casos ~ semanas_epidemiologicas + provincia_id + departamento_id + semanas_epidemiologicas + grupo_edad_id"
especifica la relación entre la variable objetivo "cantidad_casos" y las
variables predictoras "semanas_epidemiologicas", "provincia_id", "departamento_id" y "grupo_edad_id".
Estas variables se utilizan para predecir "cantidad_casos".

model_fit = model.fit(): Esta línea ajusta el modelo especificado a los datos utilizando el método .fit()
del modelo model. Después de ejecutar esta línea, model_fit contendrá los resultados del ajuste del modelo,
incluyendo los coeficientes de regresión estimados, estadísticas de ajuste y más.

print(model_fit.summary()): Finalmente, esta línea imprime un resumen estadístico del modelo ajustado
utilizando model_fit.summary(). El resumen incluirá información importante como los coeficientes, los
valores p, el estadístico R² y otros indicadores de calidad del ajuste del modelo.

Este resumen es útil para analizar la calidad y la significancia del modelo ajustado, así como para
interpretar los coeficientes de regresión. Puedes utilizar la información del resumen para evaluar
qué variables son importantes para predecir "cantidad_casos" y cómo se relacionan con la variable objetivo. """

# Fit a model to the data
model = ols(formula="cantidad_casos ~  + semanas_epidemiologicas +	provincia_id	+ departamento_id + semanas_epidemiologicas + grupo_edad_id ", data=df)
model_fit = model.fit()

print(model_fit.summary())
```

El resultado obtenido es la salida de un modelo de regresión lineal ordinaria (OLS) aplicado a los datos. Aquí hay una explicación de algunos de los puntos clave de esta salida:

1. Dep. Variable: Esto indica la variable dependiente, es decir, la que estás tratando de predecir o explicar. En este caso, es "cantidad_casos".

2. R-squared: El coeficiente de determinación R-cuadrado es una medida de cuánta variación en la variable dependiente (cantidad_casos) es explicada por las variables independientes en el modelo. En este caso, R-squared es 0.056, lo que significa que solo el 5.6% de la variación en "cantidad_casos" se explica por las variables independientes incluidas en el modelo. Esto sugiere que el modelo no se ajusta muy bien a los datos.

3. Adj. R-squared: Es una versión ajustada de R-squared que tiene en cuenta el número de predictores en el modelo. En este caso, es 0.043.

4. F-statistic: El valor F y su correspondiente p-valor indican la significancia global del modelo. En este caso, el F-statistic es 4.285 y el p-valor asociado es 0.00220, lo que sugiere que al menos una de las variables independientes es significativa en el modelo.

5. Df Residuals: Representa los grados de libertad de los residuos, es decir, el número de observaciones menos el número de parámetros estimados.

6. Df Model: Los grados de libertad del modelo son el número de parámetros estimados, en este caso, 4.

7. coef: Los coeficientes estimados para cada variable independiente en el modelo. Estos coeficientes indican cuánto cambia la variable dependiente por un cambio unitario en la variable independiente correspondiente. Por ejemplo, para "semanas_epidemiologicas", el coeficiente es -0.0323, lo que significa que un incremento de una unidad en "semanas_epidemiologicas" se asocia con una disminución de 0.0323 unidades en "cantidad_casos".

8. P>|t|: Los valores p indican la significancia estadística de cada coeficiente. Un valor p bajo (generalmente < 0.05) sugiere que el coeficiente es estadísticamente significativo. En este caso, solo el coeficiente "Intercept" es significativo (p = 0.011).

9. Omnibus, Prob(Omnibus): Estas estadísticas están relacionadas con la normalidad de los residuos. Un valor pequeño de "Prob(Omnibus)" sugiere que los residuos no se distribuyen normalmente.

10. Durbin-Watson: Evalúa la autocorrelación de los residuos. Un valor cerca de 2 indica que no hay autocorrelación. En este caso, el valor es bastante bajo, lo que sugiere la posibilidad de autocorrelación.

11. Jarque-Bera (JB): Es una prueba de la asimetría y la curtosis de los residuos. Un valor alto de JB y un valor p bajo sugieren que los residuos no siguen una distribución normal.

12. Cond. No.: La condición número es una medida de multicolinealidad. Un valor alto podría indicar problemas de multicolinealidad entre las variables independientes.

```python
""" El código importa la clase Ridge del módulo linear_model de scikit-learn.
Ridge es un tipo de modelo de regresión lineal regularizado, conocido como Regresión Ridge.

La Regresión Ridge es similar a la regresión lineal ordinaria (OLS), pero incluye
un término de regularización L2 en la función de costo. Este término de regularización
ayuda a prevenir el sobreajuste al penalizar los coeficientes del modelo, lo que a menudo
resulta en una mejor generalización del modelo. """

from sklearn.linear_model import Ridge
```

```python
""" Definimos un modelo de Regresión Ridge con un valor específico de alpha.
El parámetro alpha en la Regresión Ridge controla la fuerza de la regularización.
Un valor pequeño de alpha (como 0.0001 en este caso) significa que la regularización será débil,
lo que permite que el modelo se ajuste más a los datos de entrenamiento.

En términos generales:

* Si alpha es igual a 0, la Regresión Ridge es equivalente a la regresión lineal ordinaria (OLS).

* A medida que alpha aumenta, la fuerza de la regularización se incrementa, lo que reduce la complejidad
del modelo y ayuda a prevenir el sobreajuste.

La elección del valor de alpha depende de los datos y de nuestross objetivos. Un valor pequeño de alpha
como 0.0001 sugiere que estás minimizando la influencia de la regularización y permitiendo que el
modelo se ajuste más a los datos. (En este caso opte por un alpha de 0.1, lo cual aumento el resultado
del R_2 en 0,1. Bastante despreciable la verdad)"""

ridge = Ridge(alpha=0.1) # definimos el parámetro alpha=0.0001
```

```python
""" con la línea de código ridge.fit(X_train, y_train) ajustamos el modelo de Regresión Ridge a los datos de entrenamiento.
Esto significa que el modelo ha aprendido a relacionar las características en X_train
con las etiquetas en y_train utilizando la regularización Ridge, que es una forma de
regresión lineal regularizada."""

ridge.fit(X_train,y_train)
```

```python
""" Realizamos predicciones en el conjunto de prueba utilizando el modelo de Regresión Ridge.
Las predicciones se almacenan en la variable y_pred. Estas predicciones representan los valores
estimados por el modelo para la variable dependiente (etiquetas) con base en las
características del conjunto de prueba (X_test). """

# Predict:
y_pred = ridge.predict(X_test)
```

```python
""" En este código, comparamos modelos de Regresión Ridge con diferentes
valores de alpha en el conjunto de entrenamiento.

1. Importamos la clase Ridge de scikit-learn.

2. Definimos una lista alpha_values que contiene los diferentes valores de alpha
que queremoss evaluar. La elección de alpha afectará la fuerza de la regularización
en los modelos Ridge.

3. Inicializamos una lista vacía llamada ridge_models que se utilizará para almacenar
los modelos de Regresión Ridge entrenados con diferentes valores de alpha.

4. Iniciamos un bucle for que recorre los valores de alpha en la lista alpha_values.

5. En cada iteración del bucle, inicializamos un modelo de Regresión Ridge con el
valor de alpha actual y ajustas el modelo a los datos de entrenamiento (X_train y y_train)
utilizando ridge_model.fit(X_train, y_train).

6. Almacenamos el modelo ajustado en la lista ridge_models para su posterior uso.

7. Realizamos predicciones en el conjunto de prueba (X_test) utilizando el modelo actual y
calculamos el error cuadrático medio (MSE) entre las predicciones y las etiquetas reales.

8. Imprimimos el valor de alpha y el MSE correspondiente para cada modelo Ridge
ajustado en el conjunto de entrenamiento.

Esto permite comparar cómo se desempeñan los diferentes modelos de Regresión
Ridge con diferentes valores de alpha en el conjunto de prueba y determinar cuál
es el valor de alpha que produce el mejor rendimiento en términos de MSE en los datos. """

from sklearn.linear_model import Ridge

# valores de alpha que vamos a evaluar
alpha_values = [0.0001,0.001,0.01,0.1]
ridge_models = []

print('Model comparison on the TRAINING SET:')

for alpha in alpha_values:
  # inicializamos un modelo de Ridge regression
  ridge_model = Ridge(alpha=alpha, fit_intercept=True)
  # lo entrenamos
  ridge_model.fit(X_train, y_train)
  ridge_models.append(ridge_model)
  # y lo evaluamos sobre los datos de training utilizando MSE
  y_hat_train = ridge_model.predict(X_test)
  current_mse = mean_squared_error(y_hat_train, y_test)
  print('Ridge with alpha: {:.4f} \t MSE = {:.4f}'.format(alpha, current_mse))
```

Los resultados que obtenidos muestran la comparación del rendimiento de diferentes modelos de Regresión Ridge con diferentes valores de alpha en el conjunto de entrenamiento.

Se están evaluando cuatro modelos de Regresión Ridge con diferentes valores de alpha: 0.0001, 0.001, 0.01 y 0.1.

Cada modelo se ajusta a los datos de entrenamiento (X_train e y_train) y se evalúa utilizando el error cuadrático medio (MSE) en el conjunto de prueba (X_test e y_test).

Los resultados muestran que, a medida que el valor de alpha aumenta, el MSE disminuye. Esto significa que los modelos con un alpha más alto tienen un mejor ajuste en el conjunto de prueba, al menos en términos de MSE.

El MSE es una medida que mide la dispersión de los errores cuadrados, y un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el modelo con alpha igual a 0.1 tiene el MSE más bajo en el conjunto de prueba, lo que sugiere que es el modelo que mejor se ajusta a nuestros datos de prueba entre los valores de alpha evaluados.

Estos resultados indican que, para este conjunto de datos y problema específico, un valor de alpha de 0.1 parece ser el más adecuado en términos de ajuste y rendimiento en el conjunto de prueba, al menos según el MSE.

```python
# Definimos una función que mapea los valores de 'grupo_edad_desc' a 'target'
def asignar_target(grupo_edad_desc):
    menores = ["De 5 a 9 años", "De 2 a 4 años", "Posneonato (29 hasta 365 dias)", "De 13 a 24 meses", "De 10 a 14 años", "De 15 a 19 años"]

    if grupo_edad_desc in menores:
        return 'Menores'
    else:
        return 'Mayores'

# Aplicamos la función a la columna 'grupo_edad_desc' y crea la columna 'target'
df['target'] = df['grupo_edad_desc'].apply(lambda x: asignar_target(x))


df
```

```python
# Separar en X y y
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
# crear el procedimiento:
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
skf.get_n_splits(X, y)
```

```python
categorical_columns = ['departamento_nombre', 'provincia_nombre', 'evento_nombre', 'grupo_edad_desc']
X = X.drop(categorical_columns, axis=1)
```

```python
# Modelo vacío
model = RandomForestClassifier(random_state=1, n_estimators=30, criterion="gini", max_depth=4)

# Lista para guardar accuracy
list_ac = []

for i in range(0, skf.get_n_splits(X, y)):
    # Calculo de índices
    train, test = list(skf.split(X, y))[i]

    # Separar train y test utilizando iloc para seleccionar filas por índice entero
    X_train_n, y_train_n, X_test_n, y_test_n = X.iloc[train], y.iloc[train], X.iloc[test], y.iloc[test]

    # Ajustar el modelo
    model.fit(X_train_n, y_train_n)

    # Predicciones
    pred = model.predict(X_test_n)

    # Calculo del error
    accuracy = accuracy_score(y_true=y_test_n, y_pred=pred)
    print('Iteracion: ', i + 1, ' Accuracy:', accuracy)
    list_ac.append(accuracy)
```

"Iteracion: 1 Accuracy: 0.5254237288135594": En la primera iteración de la validación cruzada, el modelo alcanzó una precisión del 52.54%, lo que significa que aproximadamente el 52.54% de las predicciones fueron correctas en el conjunto de prueba de esa iteración.

"Iteracion: 2 Accuracy: 0.559322033898305": En la segunda iteración, la precisión del modelo fue del 55.93%, lo que indica que alrededor del 55.93% de las predicciones fueron correctas en el conjunto de prueba de esa iteración.

"Iteracion: 3 Accuracy: 0.559322033898305": En la tercera iteración, nuevamente se obtuvo una precisión del 55.93%, lo que significa que aproximadamente el 55.93% de las predicciones fueron correctas en el conjunto de prueba de esa iteración.

"Iteracion: 4 Accuracy: 0.5689655172413793": La cuarta iteración resultó en una precisión del 56.90%, lo que indica que alrededor del 56.90% de las predicciones fueron correctas en el conjunto de prueba de esa iteración.

"Iteracion: 5 Accuracy: 0.5517241379310345": En la quinta y última iteración, el modelo obtuvo una precisión del 55.17%, lo que significa que alrededor del 55.17% de las predicciones fueron correctas en el conjunto de prueba de esa iteración.

```python
print('Accuracy promedio:', round(np.mean(list_ac),2))
print('Accuracy std:', round(np.std(list_ac),2))
```
