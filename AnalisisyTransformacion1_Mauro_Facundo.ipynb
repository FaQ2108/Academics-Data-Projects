```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('informacion-publica-dengue-zika-nacional-anio-2022.csv', encoding='latin-1', sep=';')

df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Moda en semanas epidemiologicas:
df['semanas_epidemiologicas'].mode()

#La semana epidemiológica 14 es la más común en los registros de esa columna.
```

```python
# Moda en provincia_nombre:
df['provincia_nombre'].mode()

# La mayoría de los registros en esa columna corresponden a la provincia de Salta.
```

```python
# Moda en grupo_edad_desc:
df['grupo_edad_desc'].mode()

# La mayoria de los registros corresponden al intervalo de edad De 45 a 65 años.
```

```python
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12, 8))

ax = sns.regplot(x="cantidad_casos", y="semanas_epidemiologicas", data=df)
ax.set(xlabel='cantidad_de_casos', ylabel='semanas_epidemiologicas')
plt.title('Cantidad de Casos por Semana Epidemiologica')
plt.show()
```

```python
df['provincia_nombre'].value_counts()
```

```python
df['grupo_edad_desc'].value_counts()
```

```python
# Hacer la agrupacion
df_grpd = df.groupby("evento_nombre").count().reset_index()

# Grafico
fig = plt.figure(figsize=(12, 8))
sns.barplot(x="evento_nombre", y="cantidad_casos", data=df_grpd)
ax.set(xlabel='Evento', ylabel='Casos')
plt.show()

#Todos los datos parecen corresponder a casos de dengue, aunque la descripcion anexada a la base de datos,
#asi como el nombre de la misma indique que se trata de casos de dengue y zika.
```

```python
# Hacer la agrupacion
df_grpd = df.groupby("provincia_nombre").count().reset_index()

# Grafico
fig = plt.figure(figsize=(15,8))
sns.barplot(x="provincia_nombre", y="cantidad_casos", data=df_grpd)
ax.set(xlabel='Provincia', ylabel='Casos')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
sns.boxplot(x="provincia_nombre", y='cantidad_casos', data=df)
plt.title('Provincia vs Cantidad_de_Casos')
plt.show()
```
