```python

mean_y = np.mean(y)
# Calculo de la SSR (Suma de Cuadrados de la Regresión)
ssr = np.sum((np.mean(y) - y.mean())**2)

# Calculo de la SSD (Suma de Cuadrados de la Desviación)
ssd = np.sum((y - np.mean(y))**2)

# Calculo de la SST (Suma de Cuadrados Total)
sst = np.sum((y - mean_y)**2)

print("SSR:", ssr)
print("SSD:", ssd)
print("SST:", sst)
```

```python
x_1= np.linspace(stats.norm(10, 1.2).ppf(0.01),stats.norm(10, 1.2).ppf(0.99))
FDP_normal = stats.norm(10, 1.2).pdf(x_1) # FDP
plt.plot(x_1, FDP_normal, label='FDP nomal')
plt.title('Función de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
df = pd.DataFrame({
    'grupo_edad_desc': ['De 10 a 14 años', 'De 15 a 19 años', 'De 20 a 24 años', 'De 10 a 14 años', 'De 15 a 19 años'],
    'cantidad_casos': [15, 20, 10, 12, 18]
})

df.head()
```

```python
contigency= pd.crosstab(df['grupo_edad_desc'], df['cantidad_casos'])
contigency
```

```python
contigency_pct = pd.crosstab(df['grupo_edad_desc'], df['cantidad_casos'], normalize='index')
contigency_pct
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(contigency, annot=True, cmap="YlGnBu")
```

```python
from scipy.stats import chi2_contingency
```

```python
# Chi-square test of independence.
c, p, dof, expected = chi2_contingency(contigency)
# Print the p-value
print(p)
```
